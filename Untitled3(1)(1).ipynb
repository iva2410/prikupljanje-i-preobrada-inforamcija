# Čišćenje

Tekst je dodatno obrađen primenom standardnih tehnika:

- konverzija u mala slova (case folding)

- uklanjanje specijalnih karaktera

- uklanjanje stop reči

- normalizacija teksta

Cilj ove faze je smanjenje šuma i standardizacija ulaznih podataka.

```python
def normalize_text(text):

    text = text.lower()

    text = re.sub(r'<.*?>', '', text)

    text = re.sub(r'[^a-zA-Z\s]', '', text)

    text = re.sub(r'\s+', ' ', text)

    return text
```

```python
df['clean_text'] = df['Text'].apply(normalize_text)
```

# Tokenizacija

Tokenizacija predstavlja proces razlaganja teksta na manje jedinice, najčešće reči ili tokene, koji se dalje koriste kao ulaz u modele mašinskog učenja.

U ovom radu primenjene su različite vrste tokenizacije u zavisnosti od korišćenog modela. Dok jednostavna tokenizacija na nivou reči odgovara tradicionalnim metodama, napredni modeli poput BERT-a koriste složenije pristupe koji omogućavaju bolje razumevanje jezika.

Word tokenization podrazumeva podelu teksta na pojedinačne reči. Ovo je najčešće korišćena metoda tokenizacije u klasičnim NLP pristupima kao što su Bag-of-Words i TF-IDF.

### Sentence tokenization  

Kod sentence tokenization svaka recenzija se deli na pojedinačne rečenice. Ovaj pristup je koristan kod analize dužih tekstova, jer omogućava modelima da razumeju strukturu dokumenta na nivou rečenica.

```python
from nltk.tokenize import sent_tokenize

df['sentences'] = df['Text'].apply(sent_tokenize)
```

### Subword tokenization

Kod savremenih transformer modela kao što su BERT, GPT i RoBERTa, koristi se subword tokenization. Ova metoda deli reči na manje jedinice (subwords), što omogućava modelima da obrađuju retke ili nepoznate reči.

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def bert_tokenize(text):
    encoded = tokenizer(
        text,
        truncation=True,
        max_length=510,
        add_special_tokens=False
    )
    return tokenizer.convert_ids_to_tokens(encoded["input_ids"])

df['bert_tokens'] = df['clean_text'].apply(bert_tokenize)
```

# Augmentation

U cilju poboljšanja performansi modela i povećanja raznovrsnosti podataka, u radu su primenjene tehnike augmentacije teksta. Augmentacija podataka predstavlja proces generisanja novih uzoraka na osnovu postojećih, uz očuvanje osnovnog značenja i sentimenta.

Ovaj pristup je posebno značajan kod zadataka obrade prirodnog jezika, gde dostupnost velikih i balansiranih skupova podataka često predstavlja izazov.

### Zamena sinonima (Synonym Replacement)

Odabrane reči u tekstu zamenjuju se njihovim sinonimima, čime se dobija nova varijanta rečenice uz očuvanje značenja.

```python
def get_synonyms(word):

    synonyms = set()

    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonym = lemma.name().replace("_"," ")
            synonyms.add(synonym)

    return list(synonyms)
```

```python
def synonym_replacement(tokens, n=1):

    new_tokens = tokens.copy()
    random_words = list(set(tokens))
    random.shuffle(random_words)

    num_replaced = 0

    for word in random_words:

        synonyms = get_synonyms(word)

        if len(synonyms) > 0:

            synonym = random.choice(synonyms)

            new_tokens = [synonym if w == word else w for w in new_tokens]

            num_replaced += 1

        if num_replaced >= n:
            break

    return new_tokens
```

```python
import random

df['augmented_tokens'] = df['word_tokens'].apply(lambda x: synonym_replacement(x, n=1))
```

### Brisanje reči (Word Deletion)

Nasumično uklanjanje pojedinih reči iz teksta, čime se simulira skraćena ili nepotpuna rečenica.

```python
def random_deletion(tokens, p=0.2):

    if len(tokens) == 1:
        return tokens

    new_tokens = []

    for word in tokens:

        r = random.uniform(0,1)

        if r > p:
            new_tokens.append(word)

    if len(new_tokens) == 0:
        return [random.choice(tokens)]

    return new_tokens
```

```python
df['augmented_delete_tokens'] = df['word_tokens'].apply(lambda x: random_deletion(x,0.2))
```

### Zamena mesta reči (Random Swap)

Nasumična zamena pozicija dve reči u rečenici, čime se uvodi varijacija u strukturi teksta.

```python
def random_swap(words, n=1):
    words = words.copy()

    for _ in range(n):
        if len(words) < 2:
            return words

        i, j = random.sample(range(len(words)), 2)
        words[i], words[j] = words[j], words[i]

    return words

df['swap_tokens'] = df['word_tokens'].apply(lambda x: random_swap(x))
df['swap_text'] = df['swap_tokens'].apply(lambda x: " ".join(x))
```

### Brisanje rečenica (Sentence Deletion)

Uklanjanje cele rečenice iz teksta, čime se dobija kraća verzija dokumenta.

```python
def sentence_deletion(sentences):
    if not isinstance(sentences, list) or len(sentences) <= 1:
        return sentences

    temp_sentences = list(sentences)

    idx = random.randint(0, len(temp_sentences) - 1)

    temp_sentences.pop(idx)

    return temp_sentences

df['sent_deleted'] = df['sentences'].apply(sentence_deletion)

df['sent_deleted_text'] = df['sent_deleted'].apply(
    lambda x: " ".join([str(sent) for sent in x]) if isinstance(x, list) else ""
)
```

### Zamena mesta rečenica (Sentence Swap)

Promena redosleda rečenica u tekstu, čime se menja struktura bez gubitka sadržaja

```python
def sentence_swap(sentences):
    if not isinstance(sentences, list) or len(sentences) < 2:
        return sentences

    swapped_list = list(sentences)

    i, j = random.sample(range(len(swapped_list)), 2)

    swapped_list[i], swapped_list[j] = swapped_list[j], swapped_list[i]

    return swapped_list

df['sent_swapped'] = df['sentences'].apply(sentence_swap)

df['sent_swapped_text'] = df['sent_swapped'].apply(
    lambda x: " ".join([str(sent) for sent in x]) if isinstance(x, list) else ""
)
```

# Treniranje

U praktičnom delu rada trenirali smo različite modele mašinskog učenja za klasifikaciju sentimenta na Amazon recenzijama. Cilj je bio ispitati uticaj predobrade teksta (clean tokens, stemming, lematizacija, augmentacije) na performanse modela.

Koristili smo sledeće pristupe:

- Bag of Words (BoW) - Predstavlja tekst kao vektor frekvencija reči, ne uzima u obzir redosled i kontekst.
- TF-IDF - Proširenje BoW metode koje dodeljuje težine rečima na osnovu njihove važnosti u dokumentu.
- Word2Vec - Metoda zasnovana na neuronskim mrežama koja uči vektorske reprezentacije reči na osnovu njihovog konteksta.
- FastText - Proširenje Word2Vec pristupa koje koristi podreči (subword jedinice), čime bolje obrađuje retke reči i morfološke varijacije.
- BERT - Napredni model zasnovan na transformerskoj arhitekturi koji koristi kontekstualne reprezentacije reči i razume značenje teksta u zavisnosti od okruženja.

Za treniranje i evaluaciju modela korišćen je podeljeni skup podataka 80% za treniranje, 20% za testiranje.

Za modele koji koriste bag-of-words (BoW) i TF-IDF reprezentaciju teksta trenirano je nad klasifikatorima Support Vector Machine (SVM), Logistic Regression (LogReg) i Naive Bayes (NB). Da bi se balansirala klasa, korišćen je SMOTE za generisanje sintetičkih primera iz manjinske klase.

Za modele koji koriste embedding reprezentacije (Word2Vec i FastText) trenirano je samo sa SVM i Logistic Regression, jer Naive Bayes ne funkcioniše dobro sa embedding vektorima (posebno kada postoje negativne vrednosti).
Balansiranje klasa je ostvareno korišćenjem RandomOverSampler, koji nasumično duplicira instance manjinske klase dok se ne postigne balans sa većinskom klasom.

Za BERT modele trenirano je direktno nad BERTForSequenceClassification.
Balansiranje klasa je ostvareno korišćenjem WeightedRandomSampler, koji dodaje težine uzorcima pri treniranju, tako da modeli uzimaju u obzir retke klase češće. Ovo omogućava ravnomerniji doprinos svake klase u optimizaciji gubitka, a posebno je pogodno za velike modele sa batch treniranjem.

Modeli su trenirani nad različitim verzijama teksta (text_versions), uključujući word-level, sentence-level i BERT tokenizaciju.


```python
baseline_versions = {
    "Raw Text": "Text",
    "Clean Text": "clean_text"
}
text_versions = {
    # ---------- Word-level ----------
    "Clean (Word)" : "word_tokens",
    "Stemmed (Word)": "stemmed_text",
    "Lemmatized (Word)": "lemmatized_text",
    "Synonym replacement (Word)": "augmented_tokens",
    "Word Delete (Word)": "augmented_delete_text",
    "Random Swap (Word)": "swap_text",
    "Sentence Deletion (Word)": "sent_deleted_text",
    "Sentence Swap (Word)": "sent_swapped_text",

    # ---------- Sentence-level ----------
    "Clean (Sentence)": "sentences",
    "Stemmed (Sentence)": "stemmed_sentences_text",
    "Lemmatized (Sentence)": "lemmatized_sentences_text",

    # ---------- BERT / Subword ----------
    "Clean (BERT)": "bert_tokens"
}
```

```python
def prepare_text(column):
    return df[column].apply(lambda x: " ".join(x) if isinstance(x, list) else x)

def prepare_tokens(column):
    return df[column].apply(lambda x: x if isinstance(x, list) else x.split())

def flatten_tokens(x):
    if isinstance(x, str):
        return x
    elif isinstance(x, list):
        flat = []
        for item in x:
            if isinstance(item, list):
                flat.extend(item)
            else:
                flat.append(item)
        flat = [str(i) for i in flat]
        return " ".join(flat)
    else:
        return str(x)
```

```python

for name, col in {**baseline_versions, **text_versions}.items():
    df[col + "_str"] = df[col].apply(flatten_tokens)

def train_evaluate_smote(text_column, vectorizer_type="bow", model_type="nb"):

    texts = df[text_column]
    y = df['Sentiment']

    X_train, X_test, y_train, y_test = train_test_split(
        texts, y, test_size=0.2, random_state=42, stratify=y
    )

    if vectorizer_type == "bow":
        vect = CountVectorizer(max_features=5000)
    elif vectorizer_type == "tfidf":
        vect = TfidfVectorizer(max_features=5000)
    else:
        raise ValueError("Vectorizer type must be 'bow' or 'tfidf'.")

    X_train_vect = vect.fit_transform(X_train)
    X_test_vect = vect.transform(X_test)

    X_train_dense = X_train_vect.toarray()

    smote = SMOTE(random_state=42)
    X_train_res, y_train_res = smote.fit_resample(X_train_dense, y_train)

    if model_type == "nb":
        model = MultinomialNB()
    elif model_type == "logreg":
        model = LogisticRegression(max_iter=1000)
    elif model_type == "svm":
        model = LinearSVC()
    else:
        raise ValueError("Unknown model")

    model.fit(X_train_res, y_train_res)
    y_pred = model.predict(X_test_vect)

    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    return acc, precision, recall, f1

models = ["nb", "logreg", "svm"]

baseline_results = []

for name, col in baseline_versions.items():

    for model_type in models:
        acc_bow, p_bow, r_bow, f1_bow = train_evaluate_smote(col + "_str", "bow", model_type=model_type)
        acc_tfidf, p_tfidf, r_tfidf, f1_tfidf = train_evaluate_smote(col + "_str", "tfidf", model_type=model_type)

        baseline_results.append({
            "Version": name,
            "Model": model_type,
            "BoW Accuracy": acc_bow,
            "BoW Precision": p_bow,
            "BoW Recall": r_bow,
            "BoW F1": f1_bow,
            "TF-IDF Accuracy": acc_tfidf,
            "TF-IDF Precision": p_tfidf,
            "TF-IDF Recall": r_tfidf,
            "TF-IDF F1": f1_tfidf
        })

baseline_df = pd.DataFrame(baseline_results)
print("Baseline results:")
print(baseline_df)


results_smote = []

for name, col in text_versions.items():
    if "BERT" in name:
        continue
    # check if 'col' is a list of lists (like 'sentences') before adding '_str'
    # This is handled by flatten_tokens, so direct application is fine.
    col_to_process = col + "_str" if isinstance(df[col].iloc[0], list) else col

    for model_type in models:
        # Ensure we are passing the correct column for vectorization
        acc_bow, p_bow, r_bow, f1_bow = train_evaluate_smote(col + "_str", "bow", model_type=model_type)
        acc_tfidf, p_tfidf, r_tfidf, f1_tfidf = train_evaluate_smote(col + "_str", "tfidf", model_type=model_type)

        results_smote.append({
            "Version": name,
            "Model": model_type,
            "BoW Accuracy": acc_bow,
            "BoW Precision": p_bow,
            "BoW Recall": r_bow,
            "BoW F1": f1_bow,
            "TF-IDF Accuracy": acc_tfidf,
            "TF-IDF Precision": p_tfidf,
            "TF-IDF Recall": r_tfidf,
            "TF-IDF F1": f1_tfidf
        })

results_smote_df = pd.DataFrame(results_smote)
print(results_smote_df)
```

```python
def prepare_tokens_for_embeddings(col):
    tokens_list = []
    for doc in df[col]:
        if isinstance(doc, list):
            flat_doc = []
            for t in doc:
                if isinstance(t, list):
                    flat_doc.extend([str(x) for x in t])
                else:
                    flat_doc.append(str(t))
            tokens_list.append(flat_doc)
        else:
            tokens_list.append(str(doc).split())
    return tokens_list

def sentence_vector(tokens, model):
    vecs = [model.wv[word] for word in tokens if word in model.wv]
    if len(vecs) == 0:
        return np.zeros(model.vector_size)
    return np.mean(vecs, axis=0)

def train_embedding_with_ros(tokens_list, y, embedding_type="w2v", model_type="logreg"):
    if embedding_type == "w2v":
        embed_model = Word2Vec(sentences=tokens_list, vector_size=100, window=5, min_count=2, workers=4)
    elif embedding_type == "ft":
        embed_model = FastText(sentences=tokens_list, vector_size=100, window=5, min_count=2, workers=4)
    else:
        raise ValueError("embedding_type must be 'w2v' or 'ft'")

    X = np.array([sentence_vector(t, embed_model) for t in tokens_list])

    # RandomOverSampler
    ros = RandomOverSampler(random_state=42)
    X_res, y_res = ros.fit_resample(X, y)

    X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)

    if model_type == "logreg":
        clf = LogisticRegression(max_iter=1000)
    elif model_type == "svm":
        clf = LinearSVC()
    else:
        raise ValueError("model_type must be 'logreg', or 'svm'")

    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    return acc, precision, recall, f1

# Baseline
models = [ "logreg", "svm"]
embedding_types = ["w2v", "ft"]
baseline_results = []

for name, col in baseline_versions.items():
    tokens_list = prepare_tokens_for_embeddings(col)
    y = df['Sentiment'].values

    for embed in embedding_types:
        for model_type in models:
            acc, precision, recall, f1 = train_embedding_with_ros(tokens_list, y, embed, model_type)
            baseline_results.append({
                "Version": name,
                "Embedding": embed,
                "Model": model_type,
                "Accuracy": acc,
                "Precision": precision,
                "Recall": recall,
                "F1": f1
            })
baseline_df = pd.DataFrame(baseline_results)
print("Baseline embedding results:")
print(baseline_df)
results = []

for name, col in text_versions.items():
    if "BERT" in name:
        continue

    tokens_list = prepare_tokens_for_embeddings(col)
    y = df['Sentiment'].values

    for embed in embedding_types:
        for model_type in models:
            try:
                acc, precision, recall, f1 = train_embedding_with_ros(tokens_list, y, embed, model_type)
                results.append({
                    "Version": name,
                    "Embedding": embed,
                    "Model": model_type,
                    "Accuracy": acc,
                    "Precision": precision,
                    "Recall": recall,
                    "F1": f1
                })
            except Exception as e:
                print(f"Greška kod {name}, {embed}, {model_type}: {e}")

results_df = pd.DataFrame(results)
print(results_df)

```

SVM generalno daje bolje F1 i Precision rezultate, što je očekivano za embeddings jer SVM može bolje da separiše višedimenzionalne vektore u latentnom prostoru.

FastText embeddings daju nešto slabije performanse u odnosu na Word2Vec, što može biti posledica toga što FastText generiše vektore i za nepoznate reči, ali sa ovim skupom podataka možda to nije donelo dodatnu prednost.

```python
import torch
from torch.utils.data import DataLoader, TensorDataset, random_split
from transformers import BertTokenizer, BertForSequenceClassification
from torch.optim import AdamW
from torch.utils.data import DataLoader, TensorDataset, random_split, WeightedRandomSampler


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
MAX_LEN = 510

def encode_texts(texts):
    return tokenizer(
        texts.tolist(),
        add_special_tokens=True,
        max_length=MAX_LEN,
        padding='max_length',
        truncation=True,
        return_tensors='pt'
    )

def create_dataloader(text_series, labels, batch_size=16, train=True):
        train_texts, test_texts, train_labels, test_labels = train_test_split(
            text_series, labels, test_size=0.2, stratify=labels, random_state=42
        )

        train_enc = encode_texts(train_texts)
        test_enc = encode_texts(test_texts)

        train_dataset = TensorDataset(
            train_enc['input_ids'],
            train_enc['attention_mask'],
            torch.tensor(train_labels.values)
        )
        test_dataset = TensorDataset(
        test_enc['input_ids'],
        test_enc['attention_mask'],
        torch.tensor(test_labels.values)
        )


        class_counts = np.bincount(train_labels, minlength=2)
        class_weights = 1. / (class_counts + 1e-6)

        sample_weights = [class_weights[label] for label in train_labels]

        sampler = WeightedRandomSampler(
            weights=sample_weights,
            num_samples=len(sample_weights),
            replacement=True
        )

        train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=sampler)
        test_loader = DataLoader(test_dataset, batch_size=batch_size)

        return train_loader, test_loader

def train_bert(train_loader, model, optimizer, epochs=2):
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for batch in train_loader:
            input_ids, attention_mask, labels = [b.to(device) for b in batch]
            optimizer.zero_grad()
            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch+1}, Loss: {total_loss/len(train_loader):.4f}")

def evaluate_bert_full(test_loader, model):
    model.eval()
    preds, true_labels = [], []
    with torch.no_grad():
        for batch in test_loader:
            input_ids, attention_mask, labels = [b.to(device) for b in batch]
            outputs = model(input_ids, attention_mask=attention_mask)
            predictions = torch.argmax(outputs.logits, dim=-1)
            preds.extend(predictions.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

    acc = accuracy_score(true_labels, preds)
    precision = precision_score(true_labels, preds, average='weighted', zero_division=0)
    recall = recall_score(true_labels, preds, average='weighted', zero_division=0)
    f1 = f1_score(true_labels, preds, average='weighted', zero_division=0)

    print(classification_report(true_labels, preds, zero_division=0))

    return acc, precision, recall, f1

# Updated texts_versions to use Ollama augmented text
texts_versions = {
    "Clean Text": "clean_text"
}
df_sample = df.groupby('Sentiment', group_keys=False).apply(
    lambda x: x.sample(1000, random_state=42)
)

df_sample = df_sample.sample(frac=1, random_state=42).reset_index(drop=True)
results = {}

for name, column in texts_versions.items():
    print(f"Treniranje {name}...")

    texts = df_sample[column].apply(lambda x: " ".join(x) if isinstance(x, list) else x)
    labels = df_sample['Sentiment']

    train_loader, test_loader = create_dataloader(texts, labels)

    model = BertForSequenceClassification.from_pretrained(
        "bert-base-uncased",
        num_labels=2
    ).to(device)
    optimizer = AdamW(model.parameters(), lr=2e-5)

    train_bert(train_loader, model, optimizer, epochs=2)

    acc, precision, recall, f1 = evaluate_bert_full(test_loader, model)

    results[name] = {
        "Accuracy": acc,
        "Precision": precision,
        "Recall": recall,
        "F1": f1
    }

    print(f"{name} Metrics -> Accuracy: {acc:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1: {f1:.4f}\n")
print("\nFINAL RESULTS:")
for k, v in results.items():
    print(k, v)
```
